```python
import pandas as pd
```

```python
dfsales = pd.read_excel("df.xlsx")
```

```python
dfsales.info()
```

```python
dfsales["temp"]=dfsales["movieid"]
```

```python
dfsales["temp"]=dfsales["temp"].apply(lambda x : str(x)[:4])
```

```python
dfsales[dfsales["openyear"].isnull()]
```

```python
dfsales["openyear"].fillna(dfsales["temp"], inplace=True)
```

```python
dfsales["openyear"]=dfsales["openyear"].astype(int)
```

```python
dfsales["reopen"] = 0
```

```python
import numpy as np
```

```python
dfsales["reopen"] = np.where(dfsales["openyear"]<dfsales["salesyear"],1,0)
```

```python
dfsales["M Dist"] = dfsales["배급사"].apply(lambda x : x.split(",")[0])
```

```python
dfsales["genre"].info()
```

```python
dfsales["genre"]=dfsales["genre"].astype("str")
```

```python
dfsales["M genre"]=dfsales["genre"].apply(lambda x: x.split(",")[0])
```

```python
dfsales.columns
```

```python
dfsales2=dfsales.drop(['개봉일','매출액 점유율','대표국적', '국적', '배급사','opendt','genre','temp'],axis=1)
```

```python
dfsales2["매출액(천만원)"] = dfsales2["매출액"].apply(lambda x: x/10000000)
```

```python
dfsales2["매출액/스크린수(백만)"] = dfsales2["매출액"]/dfsales2["스크린수"]/1000000
dfsales2["매출액/상영횟수"] = dfsales2["매출액"]/dfsales2["상영횟수"]
dfsales2["매출액/관객수"] = dfsales2["매출액"]/dfsales2["관객수"]
```

```python
dfsales2.to_excel("rawdata.xlsx")
```
