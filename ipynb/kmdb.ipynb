```python
import pandas as pd
import requests
```

```python
data=pd.read_excel("movielist.xlsx")
movielist=data["영화명"].tolist()
len(movielist)
```

```python
movie = []
for i in range(1000,3270):
    baseurl = "http://api.koreafilm.or.kr/openapi-data2/wisenut/search_api/search_json2.jsp?collection=kmdb_new2"
    params = {
        "ServiceKey" : "KI1Y2B67JCCKV1OC7JAR",
        "title": movielist[i]
    }

    result = requests.get(baseurl, params=params)
    if result.json()["TotalCount"] ==0:
        continue
    else:
        mdata = result.json()["Data"][0]["Result"][0]

        mid = mdata["DOCID"]
        title = result.json()["KMAQuery"]
        titleEng = mdata["titleEng"]
        prodYear = mdata["prodYear"]
        director = mdata["directors"]["director"][0]["directorNm"]
        actor = mdata["actors"]["actor"][0]["actorNm"]
        nation = mdata["nation"]
        company = mdata["company"]
        runtime = mdata["runtime"]
        genre = mdata["genre"]
        use = mdata["use"]
        repRlsDate = mdata["repRlsDate"]
        rating = mdata["ratings"]["rating"][0]["ratingGrade"]
        Code = mdata["Codes"]["Code"][0]["CodeNo"]

        movie.append([mid,title,titleEng,prodYear,director,actor,nation,company,runtime,genre,use,repRlsDate,rating,Code])

dfmovie = pd.DataFrame(movie)
```
