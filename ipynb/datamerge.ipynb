```python
import pandas as pd
```

```python
###매출액 정보는 kofic에 영문, 장르명은 moviedb에
path1="moviedb.xlsx"
path2="kofic.xlsx"
```

```python
moviedb = pd.read_excel(path1)
moviedb.drop(columns="Unnamed: 0",inplace=True)
moviedb
```

```python
salesdb = pd.read_excel(path2)
salesdb.head()
```

### MERGE


```python
dfmovie = pd.merge(salesdb,moviedb,how="left",left_on="영화명",right_on="name")
'''info통해 데이터 확인
전체 행 4488, 
1. 장르별 수익 분석을 위해서 "genreAlt" or "repGenreNm"  필요
2. 로튼과의 비교를 위해서는 "ename" 필요
3. 국적비교 필요 "국적" from salesdb vs "nationAlt" or "repNationNm" from moviedb
4. 배급사 없는 경우 처리 어찌할지?

>> 배급사 없는 데이터 처리 여부 먼저 결정 > 영문명 없는케이스 결정 > 국적비교 후 일치하지 않는 케이스 확인 > 장르
'''
dfmovie.info()
```
