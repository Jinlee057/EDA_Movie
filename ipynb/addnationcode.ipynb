```python
import pandas as pd
```

```python
path1="kofic.xlsx"
path2="nationcode.xlsx"
```

```python
sales=pd.read_excel(path1)
nationcode=pd.read_excel(path2)
```

```python
sales.head()
```

```python
nationcode.drop(columns="Unnamed: 0",inplace=True)
```

```python
nationcode
```

```python
sales2=pd.DataFrame(sales,columns=['순위', '영화명', 'ycode', '개봉일', '매출액', '매출액 점유율', '관객수', '스크린수', '상영횟수',
       '대표국적', 'nationcode','국적','배급사'])
```

```python
saels3=pd.merge(sales2,nationcode,how="left",left_on="대표국적",right_on="nation")
```

```python
saels3.drop(columns="nationcode",inplace=True)
```

```python
saels3[saels3["대표국적"]!=saels3["nation"]]
```

```python
nationwcode = saels3.loc[:,["영화명","대표국적","code"]]
```

```python
nationwcode.drop_duplicates(subset=["영화명","대표국적"],inplace=True)
```
