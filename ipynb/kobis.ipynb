```python
list = []

for i in range(3000,3270):
    movielist_url = "http://www.kobis.or.kr/kobisopenapi/webservice/rest/movie/searchMovieList.json"
    params={
        "key" : "6f6191fa17cba50669ea533487626fae",
        "movieNm": movielist[i]
    }

    movieresult = requests.get(movielist_url, params=params)
    
    
    if movieresult.json()['movieListResult']['totCnt']==0:
        continue
    else:
        data=movieresult.json()["movieListResult"]["movieList"][0]
        mid=data["movieCd"]
        name=data["movieNm"]
        ename=data["movieNmEn"]
        prdtYear=data["prdtYear"]
        openDt=data["openDt"]
        nationAlt=data["nationAlt"]
        genreAlt=data["genreAlt"]
        typeNm=data["typeNm"]
        repNationNm=data["repNationNm"]
        repGenreNm=data["repGenreNm"]
        if len(data["directors"]) ==0:
            director=""
        else:
            director=data["directors"][0]["peopleNm"]
            
        list.append([mid,name,ename,prdtYear,openDt,nationAlt,genreAlt,typeNm,repNationNm,repGenreNm,director])
    print(i)
    
dfmovie6 = pd.DataFrame(list,columns=['mid','name', 'ename', 'prdtYear', 'openDt', 'nationAlt', 'genreAlt', 'typeNm', 'repNationNm', 'repGenreNm', 'director'])
dfmovie6
```
